```python
def Filter(x):
    if x<2:
        return 0
    elif x<6:
        return 1
    else:
        return 2    
    
tp_list=[col for col in test.columns if 'tp' in col]

def get_tp(df):
    for col in tp_list:
        df[col+'_new']=df[col].apply(Filter)
    return df

test=get_tp(test)
```

```python
tp_new_list = [col+'_new' for col in tp_list]
test[tp_new_list].head()
```

```python
for i in range(5):
    test['tp'+str(i+1).zfill(2)+'-'+'tp'+str(i+6).zfill(2)] = test['tp'+str(i+1).zfill(2)] - test['tp'+str(i+6).zfill(2)]


tp_diff_list=[col for col in test.columns if 'tp' and '-' in col]
tp_diff = test[tp_diff_list]
tps = abs(tp_diff).idxmax(axis=1)
tp_rep = []
for i in range(len(test)):
    if tps[i] == 'tp01-tp06':
        if test[tps[i][0:4]][i] > test[tps[i][5:9]][i]:
            tp_rep.append(tps[i][0:4])
        else:
            tp_rep.append(tps[i][5:9])
            
    elif tps[i] == 'tp02-tp07':
        if test[tps[i][0:4]][i] > test[tps[i][5:9]][i]:
            tp_rep.append(tps[i][0:4])
        else:
            tp_rep.append(tps[i][5:9])
            
    elif tps[i] == 'tp03-tp08':
        if test[tps[i][0:4]][i] > test[tps[i][5:9]][i]:
            tp_rep.append(tps[i][0:4])
        else:
            tp_rep.append(tps[i][5:9])
            
    elif tps[i] == 'tp04-tp09':
        if test[tps[i][0:4]][i] > test[tps[i][5:9]][i]:
            tp_rep.append(tps[i][0:4])
        else:
            tp_rep.append(tps[i][5:9])
            
    else:
        if test[tps[i][0:4]][i] > test[tps[i][5:9]][i]:
            tp_rep.append(tps[i][0:4])
        else:
            tp_rep.append(tps[i][5:9])

test['tp'] = tp_rep
test[['tp']]
```
