```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 불러오기
train = pd.read_csv('https://raw.githubusercontent.com/Chachaeul/TeamProject/master/open%20data/train.csv', index_col = [0])
train=train.reset_index(drop=True)
test = pd.read_csv('https://raw.githubusercontent.com/Chachaeul/TeamProject/master/open%20data/test_x.csv', index_col = [0])
test=test.reset_index(drop=True)
pd.set_option('display.max_columns', 150)
```

```python
train['age4'] = np.where(train['age_group'] == "10s",0,train['age_group'])
train['age4'] = np.where((train['age_group'] == "20s")|(train['age_group'] == "30s"),1,train['age4'])
train['age4'] = np.where((train['age_group'] == "40s")|(train['age_group'] == "50s"),2,train['age4'])
train['age4'] = np.where((train['age_group'] == "60s")|(train['age_group'] == "+70s"),3,train['age4'])
```

```python
del train['age_group']
del train['gender']
del train['race']
del train['religion']

del test['age_group']
del test['gender']
del test['race']
del test['religion']
```

```python
train_x = train.drop('voted', axis = 1)
train_y = train['voted']
```

```python
traincol_list = [col for col in train_x.columns if '' in col]
```

```python
testcol_list = [col for col in test.columns if '' in col]
```

## 1. StandardScaler
    - 평균을 제거하고 데이터를 단위 분산으로 조정한다. 그러나 이상치가 있다면 평균과 표준편차에 영향을 미쳐 변환된 데이터의 확산은 매우 달라지게 된다.

```python
from sklearn.preprocessing import StandardScaler
```

```python

standardScaler = StandardScaler()
print(standardScaler.fit(train_x))
train_x_ss = standardScaler.transform(train_x)
```

```python
train_x_ss = pd.DataFrame(train_x_ss)
train_x_ss.columns = traincol_list
```

```python

standardScaler = StandardScaler()
print(standardScaler.fit(test))
test_ss = standardScaler.transform(test)
```

```python
test_ss = pd.DataFrame(test_ss)
test_ss.columns = testcol_list
```

```python
test_ss.head()
```
